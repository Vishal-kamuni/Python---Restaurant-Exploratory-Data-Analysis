###2.Calculate the percentage of restaurants that serve each of the top cuisines

```python
from collections import Counter

# Drop missing and split cuisines
cuisine_series = df['Cuisines'].dropna()
all_cuisines = [c.strip() for sublist in cuisine_series.str.split(',') for c in sublist]

# Count total restaurants (with at least one cuisine)
total_restaurants = len(cuisine_series)

# Get top 3 cuisines
top_cuisines = Counter(all_cuisines).most_common(3)

# Calculate percentage
top_cuisine_percent = [(cuisine, count, round((count / total_restaurants) * 100, 2)) for cuisine, count in top_cuisines]

# Convert to DataFrame
top_cuisine_df = pd.DataFrame(top_cuisine_percent, columns=["Cuisine", "Count", "Percentage"])

# Display
print(top_cuisine_df)
```

###1. Analyze the text reviews to identify the most common positive and negative keywords.

```python
df
```

```python
!pip install nltk
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')

# Step 1: Import required libraries
import pandas as pd
import random
import string
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

# If you haven't downloaded NLTK stopwords and tokenizer before, uncomment these:
# nltk.download('punkt')
# nltk.download('stopwords')

# Step 2: Define positive, neutral, and negative review templates
positive_reviews = [
    "Amazing food and great service. Highly recommend!",
    "Excellent ambiance and delicious menu.",
    "Loved everything about this place!",
    "Tasty food, friendly staff, and quick service.",
    "A fantastic dining experience overall."
]

neutral_reviews = [
    "The food was okay, nothing special.",
    "Service was average, and so was the food.",
    "Just another decent restaurant.",
    "Had a basic experience here.",
    "Reasonably priced and okayish taste."
]

negative_reviews = [
    "Terrible service and cold food.",
    "Very disappointing experience.",
    "Not worth the money.",
    "Poor hygiene and rude staff.",
    "Won’t recommend this place."
]

# Step 3: Function to generate reviews based on rating
def generate_review(rating):
    if rating >= 4.0:
        return random.choice(positive_reviews)
    elif rating >= 2.5:
        return random.choice(neutral_reviews)
    else:
        return random.choice(negative_reviews)

# Step 4: Create a simulated 'reviews' column based on 'Aggregate rating'
df['reviews'] = df['Aggregate rating'].apply(generate_review)

# Step 5: Preprocess reviews - lowercase + remove punctuation
def preprocess_text(text):
    text = text.lower()
    text = ''.join([c for c in text if c not in string.punctuation])  # remove punctuation
    return text

df['processed_review'] = df['reviews'].astype(str).apply(preprocess_text)

# Step 6: Tokenize the cleaned review text
def tokenize_text(text):
    return word_tokenize(text)

df['tokens'] = df['processed_review'].apply(tokenize_text)



# Step 7: Remove Stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['filtered_tokens'] = df['tokens'].apply(remove_stopwords)

# Step 8: Separate Positive and Negative Reviews based on Aggregate Rating
positive_reviews_df = df[df['Aggregate rating'] >= 4.0]['filtered_tokens']
negative_reviews_df = df[df['Aggregate rating'] < 2.5]['filtered_tokens']

# Step 9: Count word frequencies
positive_word_counts = Counter()
for review in positive_reviews_df:
    positive_word_counts.update(review)

negative_word_counts = Counter()
for review in negative_reviews_df:
    negative_word_counts.update(review)

# Step 10: Display Top Positive and Negative Keywords
n = 10  # Top 10 words

print("✅ Most Common Positive Keywords:")
for word, count in positive_word_counts.most_common(n):
    print(f"{word}: {count}")

print("\n✅ Most Common Negative Keywords:")
for word, count in negative_word_counts.most_common(n):
    print(f"{word}: {count}")
```

#Task: Price Range vs. Online Delivery and Table Booking

###1. Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

```python
# Step 1: Create a summary table (cross-tab)
delivery_vs_price = pd.crosstab(df['Price range'], df['Has Online delivery'], normalize='index') * 100
booking_vs_price = pd.crosstab(df['Price range'], df['Has Table booking'], normalize='index') * 100

print("📦 Online Delivery by Price Range (%):\n", delivery_vs_price)
print("\n🍽️ Table Booking by Price Range (%):\n", booking_vs_price)
```

```python
#Online Delivery by Price Range
delivery_vs_price.rename(columns={0: 'No', 1: 'Yes'}, inplace=True)
delivery_vs_price.plot(kind='bar', stacked=True, color=['salmon', 'mediumseagreen'], figsize=(8,6))
plt.title('Online Delivery Availability by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.legend(title='Online Delivery')
plt.tight_layout()
plt.show()
```

```python
#Table Booking by Price Range
booking_vs_price.rename(columns={0: 'No', 1: 'Yes'}, inplace=True)
booking_vs_price.plot(kind='bar', stacked=True, color=['lightskyblue', 'darkblue'], figsize=(8,5))
plt.title('Table Booking Availability by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.legend(title=' Table Booking ')
plt.tight_layout()
plt.show()
```

###2.Determine if higher-priced restaurants are more likely to offer these services.

```python
# Step 1: Create a new column to label price category
df['price_category'] = df['Price range'].apply(lambda x: 'Low' if x <= 2 else 'High')

# Step 2: Calculate % of restaurants offering each service
online_delivery_by_price = pd.crosstab(df['price_category'], df['Has Table booking'], normalize='index') * 100
table_booking_by_price = pd.crosstab(df['price_category'], df['Has Online delivery'], normalize='index') * 100

# Rename for clarity
online_delivery_by_price.columns = ['No Online Delivery', 'Offers Online Delivery']
table_booking_by_price.columns = ['No Table Booking', 'Offers Table Booking']

print("📦 Online Delivery by Price Category:\n", online_delivery_by_price)
print("\n🍽️ Table Booking by Price Category:\n", table_booking_by_price)
```

```python
# Plot Table Booking
table_booking_by_price.plot(kind='bar', stacked=True, color=['lightblue', 'darkblue'], title='Table Booking by Price Category')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()

# Plot Online Delivery
online_delivery_by_price.plot(kind='bar', stacked=True, color=['lightcoral', 'mediumseagreen'], title='Online Delivery by Price Category')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```
